```python
dataset.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
important_columns = [
    'Flow Duration', 'Flow Bytes/s', 'Flow Packets/s', 'Total Fwd Packets',
    'Total Backward Packets', 'Total Length of Fwd Packets', 'Total Length of Bwd Packets',
    'Fwd Packet Length Mean', 'Bwd Packet Length Mean', 'Max Packet Length',
    'Min Packet Length', 'Packet Length Std', 'Flow IAT Mean', 'Flow IAT Std',
    'Flow IAT Max', 'Flow IAT Min', 'SYN Flag Count', 'ACK Flag Count',
    'RST Flag Count', 'PSH Flag Count', 'FIN Flag Count', 'URG Flag Count',
    'Subflow Fwd Packets', 'Subflow Bwd Packets', 'Subflow Fwd Bytes',
    'Subflow Bwd Bytes', 'Init_Win_bytes_forward', 'Init_Win_bytes_backward',
    'Active Mean', 'Active Std', 'Idle Mean', 'Idle Std', 'Label'
]
```

```python
new_final_dataset=final_dataset.sample(n=800000,random_state=42)
```

```python
augmented_dataset=pd.read_csv("augmented_data_200000.csv")
```

```python
augmented_dataset.shape
```

```python
from sklearn.utils import shuffle

merged_dataset = pd.concat([new_final_dataset, augmented_dataset], ignore_index=True)
merged_dataset = shuffle(merged_dataset, random_state=75)
```

```python
percentage_of_LDAP = ( (merged_dataset['Label'] == 'LDAP').sum() / len(merged_dataset) ) * 100
percentage_of_LDAP
```

```python
percentage_of_LDAP = ( (merged_dataset['Label'] == 'BENIGN').sum() / len(merged_dataset) ) * 100
percentage_of_LDAP
```

```python
max_value = merged_dataset['Flow Bytes/s'][~np.isinf(merged_dataset['Flow Bytes/s'])].max()
min_value = merged_dataset['Flow Bytes/s'][~np.isinf(merged_dataset['Flow Bytes/s'])].min()

merged_dataset['Flow Bytes/s'].replace(np.inf, max_value, inplace=True)
merged_dataset['Flow Bytes/s'].replace(-np.inf, min_value, inplace=True)

# Convert column to float
merged_dataset['Flow Bytes/s'] = merged_dataset['Flow Bytes/s'].astype(float)
```

```python
max_value = merged_dataset['Flow Packets/s'][~np.isinf(merged_dataset['Flow Packets/s'])].max()
min_value = merged_dataset['Flow Packets/s'][~np.isinf(merged_dataset['Flow Packets/s'])].min()

merged_dataset['Flow Packets/s'].replace(np.inf, max_value, inplace=True)
merged_dataset['Flow Packets/s'].replace(-np.inf, min_value, inplace=True)

# Convert to float
merged_dataset['Flow Packets/s'] = merged_dataset['Flow Packets/s'].astype(float)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
merged_dataset['Label']=le.fit_transform(merged_dataset['Label'])
```

```python
sorted(merged_dataset['Label'].unique())
```

```python
x=merged_dataset.drop('Label',axis=1)
```

```python
y=merged_dataset['Label']
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
import xgboost as xgb
```

```python
# Initialize the XGBoost classifier for multiclass classification
xgb_model = xgb.XGBClassifier(objective='multi:softmax', num_class=3, eval_metric='mlogloss', use_label_encoder=False)

# Train the model
xgb_model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
# Predict on test data
y_pred = xgb_model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(Y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
cm = confusion_matrix(Y_test, y_pred)
print("Confusion Matrix:")
print(cm)
```

```python
from sklearn.metrics import classification_report

report = classification_report(Y_test, y_pred)
print(report)
```
