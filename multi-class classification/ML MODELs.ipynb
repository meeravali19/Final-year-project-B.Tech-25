```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', None)
```

```python
dataset=pd.read_csv('final_dataset_merged.csv')
dataset.head(5)
```

```python
sorted(dataset['Label'].unique())
```

```python
x=max(dataset['Flow Bytes/s'])
print(x)
inf_rows = dataset[np.isinf(dataset['Flow Bytes/s'])]
inf_rows
```

```python
max_value = dataset['Flow Bytes/s'][~np.isinf(dataset['Flow Bytes/s'])].max()
min_value = dataset['Flow Bytes/s'][~np.isinf(dataset['Flow Bytes/s'])].min()

dataset['Flow Bytes/s'].replace(np.inf, max_value, inplace=True)
dataset['Flow Bytes/s'].replace(-np.inf, min_value, inplace=True)

# Convert column to float
dataset['Flow Bytes/s'] = dataset['Flow Bytes/s'].astype(float)
```

```python
max_value = dataset['Flow Packets/s'][~np.isinf(dataset['Flow Packets/s'])].max()
min_value = dataset['Flow Packets/s'][~np.isinf(dataset['Flow Packets/s'])].min()

dataset['Flow Packets/s'].replace(np.inf, max_value, inplace=True)
dataset['Flow Packets/s'].replace(-np.inf, min_value, inplace=True)

# Convert to float
dataset['Flow Packets/s'] = dataset['Flow Packets/s'].astype(float)
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dataset['Label']=le.fit_transform(dataset['Label'])
```

```python
sorted(dataset['Label'].unique())
```

data spliting

```python
x=dataset.drop('Label',axis=1)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

Model building(Logistic regression)

```python
from sklearn.linear_model import LogisticRegression
```

```python
Model=LogisticRegression()
```

```python
Model.fit(X_train,Y_train)
```

```python
y_pred=Model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy = accuracy_score(Y_test, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")
```

```python
cm = confusion_matrix(Y_test, y_pred)
print("Confusion Matrix:")
print(cm)
```

```python
from sklearn.metrics import classification_report

report = classification_report(Y_test, y_pred)
print(report)
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix

# Compute confusion matrix
cm = confusion_matrix(Y_test, y_pred)
num_classes = cm.shape[0]  # Number of classes

# Initialize lists
TP = np.diag(cm)  # True Positives (diagonal values)
FP = np.sum(cm, axis=0) - TP  # Column sum minus TP
FN = np.sum(cm, axis=1) - TP  # Row sum minus TP
TN = np.sum(cm) - (TP + FP + FN)  # Everything else

# Print per-class metrics
for i in range(num_classes):
    print(f"Class {i}:")
    print(f"  True Positives (TP): {TP[i]}")
    print(f"  False Positives (FP): {FP[i]}")
    print(f"  False Negatives (FN): {FN[i]}")
    print(f"  True Negatives (TN): {TN[i]}")
    print("-" * 30)
```
